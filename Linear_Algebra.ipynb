```python
import tensorflow as tf
```

```python
x1 = tf.constant(
    [[1, 2, 0],
     [3, 5, -1]]
)

x2 = tf.constant(
    [[37, -3, 0],
     [3, 5, 7],
     [4,5,6]]
)
```

```python
tf.linalg.matmul(x1, x2)
```

```python
print(x1.shape, x2.shape)
```

x1@x2   // Matrix multiplication

```python
x1 @ x2
```

```python
tf.linalg.matmul(
    x1,
    x2,
    transpose_a=False,
    transpose_b=True,
    adjoint_a=False,
    adjoint_b=False,
    a_is_sparse=False,
    b_is_sparse=False,
    output_type=None,
    grad_a=False,
    grad_b=False,
    name=None
)
```

```python
## Lower Band
x1_lower_band = tf.linalg.band_part(x1, 0, 1)
print(x1_lower_band)
```

```python
## Upper band
x1_upper_band = tf.linalg.band_part(x1, -1, 0)
print(x1_upper_band)
```

```python
x2 = tf.constant(
    [[37, -3, 0],
     [3, 5, 7],
     [4,5,6]], dtype = float
)
print(x2)
x2_inv = tf.linalg.inv(x2)
print(x2_inv)
```

```python
x2_identity = x2 @ x2_inv
print(x2_identity)
```

SVD


```python
s,v,d = tf.linalg.svd(x2)
print(s)
print(v)
print(d)
```
